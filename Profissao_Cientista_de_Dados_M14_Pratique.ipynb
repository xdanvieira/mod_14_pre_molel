# 2 - Esse exercício faremos por etapas:
A) Verifique se temos colunas com dados faltantes e traga a % dos dados faltantes por coluna.

```python
#seu código aqui

# Verificar dados faltantes
dados_faltantes = df.isnull().sum()

# Calcular a porcentagem de dados faltantes por coluna
porcentagem_faltantes = (dados_faltantes / len(df)) * 100

# Exibir o resultado
print("Dados faltantes por coluna:")
print(dados_faltantes)
print("\nPorcentagem de dados faltantes por coluna:")
print(porcentagem_faltantes)
```

B) Exclua nessa etapa as linhas das colunas com dados faltantes que você analisar que esse método é o ideal. Justifique o motivo de estar excluindo essas linhas.

```python
#seu código aqui

# Remover linhas com dados faltantes na coluna Genero
df.dropna(subset=['Genero'], inplace=True)

# Verificar se ainda há dados faltantes na coluna Genero
print("Dados faltantes na coluna Genero após exclusão:")
print(df['Genero'].isnull().sum())
```

JUSTIFICATIVA PARA EXCLUSÃO:

A remoção das linhas com dados faltantes na coluna Genero é uma abordagem válida porque a quantidade de dados faltantes é pequena (12 linhas), é categórica e preserva a integridade dos dados para análises ou modelagens futuras.

C) Nessa etapa substitua pela média, mediana ou moda as linhas das colunas com valores faltantes que você analisou e entendeu que esse era o melhor método. Não se esqueça de justificar sua escolha.

**Dica variáveis numéricas:** Não se esqueça de verificar a distrbuição dos dados, média e mediana.

**Dica variáveis categóricas:** Caso precise substituir uma variável categórica utilize um gráfico de barras para verificar a que mais aparece e utilizar ela na substituição.
Ou você pode optar por excluir a coluna, mas justifique sua escolha.

Caso opte por substituir você pode utilizar a função fillna:
df['SUACOLUNA'].fillna("VALOR QUE SUBSTITUIRA O NULO", inplace=True)
Ela funcionará como a replace.

```python
#seu código aqui

# Verificar estatísticas descritivas das colunas numéricas
print(df[['Pagamento_Mensal', 'Total_Pago']].describe())

# Verificar a distribuição dos dados (histograma)
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(df['Pagamento_Mensal'].dropna(), bins=30, color='blue', edgecolor='black')
plt.title('Distribuição de Pagamento_Mensal')

plt.subplot(1, 2, 2)
plt.hist(df['Total_Pago'].dropna(), bins=30, color='green', edgecolor='black')
plt.title('Distribuição de Total_Pago')

plt.show()


# Substituir valores faltantes em Pagamento_Mensal e Total_Pago pela mediana
df['Pagamento_Mensal'] = df['Pagamento_Mensal'].fillna(df['Pagamento_Mensal'].median())
df['Total_Pago'] = df['Total_Pago'].fillna(df['Total_Pago'].median())

# Verificar se ainda há dados faltantes
print("Dados faltantes após substituição:")
print(df[['Pagamento_Mensal', 'Total_Pago']].isnull().sum())
```

# 3 - Verifique se encontramos valores digitados incorretamente, ou com letras maiusculas ou minusculas, ou até mesmo mesmos valores porém escritos de forma dirente e corrija.

```python
#seu código aqui

# Padronizar os valores de gênero
df['Genero'] = df['Genero'].replace({'F': 'Female', 'f': 'Female', 'M': 'Male'})

# Padronizar os valores de serviço de internet
df['Servico_Internet'] = df['Servico_Internet'].replace({'dsl': 'DSL', 'No internet service': 'No'})

# Preencher valores em branco na coluna "Churn" com "No"
df['Churn'] = df['Churn'].fillna('No')

# Verificar as primeiras linhas do DataFrame
print(df.head())

# Verificar valores únicos na coluna "Genero"
print(df['Genero'].unique())

# Verificar valores únicos na coluna "Servico_Internet"
print(df['Servico_Internet'].unique())

# Verificar valores únicos na coluna "Churn"
print(df['Churn'].unique())

```
